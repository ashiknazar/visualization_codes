```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [5, 7, 3, 8]

plt.bar(x, y)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [5, 7, 3, 8]

plt.barh(x, y)
plt.title('Horizontal Bar Plot')
plt.xlabel('Value')
plt.ylabel('Category')
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)  # Random data from normal distribution

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sizes = [25, 30, 45]
labels = ['Category A', 'Category B', 'Category C']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
data = np.random.randn(100)

plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.plot(x, y, color="Slateblue", alpha=0.6)
plt.title('Area Plot')
plt.show()
```

```python
categories = ['A', 'B', 'C']
values1 = [5, 7, 3]
values2 = [4, 6, 8]

plt.bar(categories, values1, label='Category 1')
plt.bar(categories, values2, bottom=values1, label='Category 2')

plt.title('Stacked Bar Plot')
plt.ylabel('Value')
plt.legend()
plt.show()
```
