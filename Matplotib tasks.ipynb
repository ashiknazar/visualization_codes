# Bar Plot

```python
# Bar Plot
import matplotlib.pyplot as plt
import numpy as np

grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]

plt.bar(grades,test1)
plt.show()
```

# Giving title and label

```python
grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]

plt.bar(grades,test1)
plt.title('BAR PLOT')
plt.xlabel('Grades')
plt.ylabel('No of students')
plt.show()
```

```python
#giving fontcolour,fontsizes
grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]
plt.bar(grades,test1)
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.show()
```

# giving bar width and alignment

```python
grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]
plt.bar(grades,test1,color='lime',edgecolor='red',width=0.5,align="center")
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.show()
```

# linewidth and linestyle

```python
grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]
plt.bar(grades,test1,color='lime',edgecolor='red',width=0.5,align="center",linestyle="--",linewidth=6)
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.show()
```

# three bar

```python
grades=["A","B","C","D"]
test1=[3,10,40,35]
test2=[40,50,30,25]
test3=[35,55,60,75]

index=np.arange(len(grades))

plt.bar(index,test1,width=0.2,align='center',color='red',edgecolor="blue",linewidth=2,label='test1')
plt.bar(index+0.2,test2,width=0.2,align='center',color='lime',edgecolor="k",linewidth=2,label='test2')
plt.bar(index-0.2,test3,width=0.2,align='center',color='cyan',edgecolor="k",linewidth=2,label='test3')
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.legend()
plt.show()
print(len(grades))
```

# Draw a basic piechart

```python
import matplotlib.pyplot as plt
import numpy as np

movies=["comedy","action","romance","drama"]

votes=[2,4,3,7]

plt.pie(votes)
plt.show()
```

# Give the title, label also change label distance and legend

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1.2)
plt.legend()
plt.show()
```

# explode the values

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1,explode=[0.2,0,0.3,0])  #explode used to seperate the wedges
plt.legend(title='Genres')
plt.axis('equal')  #to seperaate the legends from center to equal
plt.show()
```

# rotating the pie chart

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1,startangle=45)  #explode used to seperate the wedges
plt.legend(title='Genres')
plt.axis('equal')  #to seperaate the legends from center to equal
plt.show()
```

# see the color of each wedge

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1,startangle=45,radius=30,colors=['lime','red','yellow','violet'])  #explode used to seperate the wedges
plt.legend(title='Genres',loc='lower left')
plt.axis('scaled')  #to seperaate the legends from center to equal
plt.show()
```

# Display percentage

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1,startangle=45,radius=30,colors=['lime','red','yellow','violet'],
       autopct='%0.3f%%',pctdistance=0.5)  #explode used to seperate the wedges
#autopct attribute is used to specify the distance of the text

#in the slice from the center of the circle

#The ratio b/w the center of each pie slice and the start of the text generated by autopct
plt.legend(title='Genres',loc='upper right')
plt.axis('scaled')  #to seperaate the legends from center to equal
plt.show()
```

# Set the text properties

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1,startangle=45,radius=30,colors=['lime','red','yellow','violet'],
       autopct='%0.1f%%',pctdistance=0.5,textprops={'fontsize':'15'})  #pass the text properties as dictionary
plt.legend(title='Genres',loc='upper right')
plt.axis('scaled')  #to seperaate the legends from center to equal
plt.show()
```

# Give the wedge properties

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1,startangle=45,radius=30,colors=['lime','red','yellow','violet'],
       autopct='%0.1f%%',pctdistance=0.5,textprops={'fontsize':'15'},
       wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'-.'})  #pass the text properties as dictionary
plt.legend(title='Genres',loc=5)
plt.axis('scaled')  #to seperaate the legends from center to equal
plt.show()
```

# giving the shadow

```python
plt.title("Favorite types of movies")
plt.pie(votes,labels=movies,labeldistance=1,startangle=45,radius=30,colors=['lime','red','yellow','violet'],
       autopct='%0.1f%%',pctdistance=0.5,textprops={'fontsize':'15'},
       wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'-.'},shadow=True)  #pass the text properties as dictionary
plt.legend(title='Genres',loc=5)
plt.axis('scaled')  #to seperaate the legends from center to equal
plt.show()
```
