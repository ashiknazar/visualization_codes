```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,2*np.pi,100)
y=np.cos(x)
fig,ax=plt.subplots()
ax.plot(x,y)
fig.savefig("figure.pdf")
plt.show()
```

```python
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
plt.show()      
```
