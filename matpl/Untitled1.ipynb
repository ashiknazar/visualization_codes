```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot the sine wave on the axes
ax.plot(x, y)

# Add title and labels
ax.set_title("Sine Wave")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")

# Add a grid
ax.grid(True)

# Display the plot
plt.show()
```
