```python
import numpy as np
import math

x=np.arange(0,math.pi*2,.05)
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.3,0.3,0.3,0.3])
y=np.sin(x)
ax1.plot(x,y,'m')
ax2.plot(x,np.cos(x),'c')
ax1.set_title('SIN ANGLE')
ax2.set_title('COS ANGLE')
plt.show()
```

```python
x=np.arange(0,math.pi*2,.05)
fig=plt.figure()
ax1=fig.add_axes([0.3,0.3,0.7,0.7])
ax2=fig.add_axes([0.6,0.6,0.4,0.4])
y=np.cos(x)
ax1.plot(x,y,'c')
ax2.plot(x,np.sin(x),'g')
ax1.set_title('COS ANGLE')
ax2.set_title('SIN ANGLE')
plt.show()
```

```python
fig,a=plt.subplots(2,3)
x=np.arange(20)
a[0][0].plot(x,x*x)
a[0][0].set_title('SQUARE')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('SQRT')
a[0][2].plot(x,x**3)
a[0][2].set_title('CUBE')
a[1][0].plot(x,np.cbrt(x))
a[1][0].set_title('CBRT')
a[1][1].plot(x,np.sqrt(x))
a[1][1].set_title('SQRTs')
a[1][2].plot(x,x*2)
a[1][2].set_title('MULTIPLE')
plt.show()
```

```python
fig,axes=plt.subplots(1,3,figsize=(10,10))
x=np.arange(10)
axes[0].plot(x,np.exp(x),'c',lw=3)
axes[0].plot(x,x*x,'m')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[0].set_title('GRAPH1')
axes[1].plot(x,np.log(x),'r')
axes[1].plot(x,x**3,'b')
axes[1].set_yscale('log')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')
axes[1].set_title('GRAPH2')
axes[2].plot(x,x,'g')
axes[2].plot(x,x*2,'k')
axes[2].set_xlabel('X axis')
axes[2].set_ylabel('Y axis')
axes[2].set_title('GRAPH3')
plt.show()
```
