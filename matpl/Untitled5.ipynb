```python

sns.kdeplot(data,x='Weight',Fill=True)  
plt.show()
```

```python
sns.jointplot(x='Height',y='Weight',data=data,dropna=True)   #scatter plot and marginal histogram, dropna= drop the null values
```

```python
sns.pairplot(data,dropna=True)
```

```python
sns.jointplot(x='Height',y='Weight',data=data,dropna=True,xlim=(0,260),ylim=(0,300))   #scatter plot and marginal histogram, dropna= drop the null values
```

```python
sns.pairplot(data,dropna=True,hue='Medal')
```

```python

corr = data['Height'].corr(data['Weight'])

sns.heatmap(corr,annot=True)
```

```python

corr_matrix = data[['Height', 'Weight']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(15,5))
# Multiple KDE plots on one graph
sns.kdeplot(data['Weight'].dropna(), color="r", label="Weight")
sns.kdeplot(data['Height'].dropna(), color="g", label="Height")
ax.set_xlabel("weight height")
plt.legend()

```
